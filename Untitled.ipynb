```python
import pandas as pd
import numpy as np
```

```python
out = pd.read_csv("./fi_out", sep="\t", header=None)
dev = pd.read_csv("./fi-uncovered", sep="\t", header=None)
```

```python
lemmas = out[1]
got = out[0]
expected = dev[0]
categories = dev[2]

results = pd.DataFrame({"lemma": lemmas, "got": got, "expected": expected, "categories": categories, "is_error": [False]*len(lemmas)})
```

```python
results.loc[results.got!=results["expected"], "is_error"] = True
errors = results[results["got"] != results["expected"]]
errors
```

```python
# From lab 2 files (with python3 adaptation)
def distance(str1, str2):
    """Simple Levenshtein implementation for evalm."""
    m = np.zeros([len(str2)+1, len(str1)+1])
    for x in range(1, len(str2) + 1):
        m[x][0] = m[x-1][0] + 1
    for y in range(1, len(str1) + 1):
        m[0][y] = m[0][y-1] + 1
    for x in range(1, len(str2) + 1):
        for y in range(1, len(str1) + 1):
            if str1[y-1] == str2[x-1]:
                dg = 0
            else:
                dg = 1
            m[x][y] = min(m[x-1][y] + 1, m[x][y-1] + 1, m[x-1][y-1] + dg)
    return int(m[len(str2)][len(str1)])

distances = []
normalized_distances = []
for i,row in errors.iterrows():
    distances.append(distance(row["expected"], row["got"]))
    normalized_distances.append(distance(row["expected"], row["got"]) / len(row["expected"]))

errors["distance"] = distances
errors["normalized_distance"] = normalized_distances
```

```python
errors.sort_values("normalized_distance", ascending=True)
```

```python
def getErrorContext(error, depth=2):
    errorIndex = errors.index[errors["expected"] == error["expected"]][0]
    if errorIndex - depth < 0 or errorIndex + depth > len(results) - 1:
        return None
    return results.iloc[errorIndex - depth: errorIndex + depth + 1]
```

```python
getErrorContext(errors.iloc[0], depth=3)
```

```python
def getSimilarContext(error, depth=2):
    context = getErrorContext(error, depth=depth).reset_index()
#     if context == None:
#         return None
    matchingContext = []
    for i in range(len(results) - ((2*depth)+1)):
        window = results.iloc[i:i+(2*depth)+1]
        windowMatches = True
        for k,row in context.iterrows():
            if row["categories"] != window.iloc[k]["categories"] and k != (len(window) // 2) + 1:
                windowMatches = False
        if windowMatches:
            matchingContext.append(window)
    return matchingContext
            
```

```python
contexts = []
done_count = 0
for i,error in errors.iterrows():
    contexts.append(getSimilarContext(error, depth=1))
    done_count+=1
    print(done_count,"/",len(errors))
```
